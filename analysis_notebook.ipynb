### Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression, Ridge

def caps_str(x):
    arr_x = [s.lower().capitalize().strip() for s in x.split(" ")]
    new_x = " ".join(arr_x)
    return new_x

educ_df = pd.read_excel("./data/bps-od_15045_indeks_pendidikan_berdasarkan_kabupatenkota_data.xlsx")
educ_df["nama_kabupaten_kota"] = educ_df["nama_kabupaten_kota"].apply(caps_str).apply(lambda x: x.replace("Kabupaten", "Kab."))
educ_df = educ_df.loc[:, "kode_kabupaten_kota":]
poverty = pd.read_excel("./data/bps-od_persentase_penduduk_miskin_berdasarkan_kabupatenkota_data.xlsx")
poverty["nama_kabupaten_kota"] = poverty["nama_kabupaten_kota"].apply(caps_str).apply(lambda x: x.replace("Kabupaten", "Kab."))
poverty = poverty.loc[:, "kode_kabupaten_kota":]

# create unique keys
educ_df["kode_kota_tahun"] = educ_df["kode_kabupaten_kota"].astype(str).str.cat(educ_df["tahun"].astype(str))
poverty["kode_kota_tahun"] = poverty["kode_kabupaten_kota"].astype(str).str.cat(poverty["tahun"].astype(str))
educ_df = educ_df.set_index("kode_kota_tahun")
poverty = poverty.set_index("kode_kota_tahun")
edu_pov = educ_df.join(poverty, how="inner", lsuffix="_edu", rsuffix="_pov")
edu_pov.head(3)
```

For the data preparation step, we created a unique keys for each dataframe, comprising of the city code and the year. After creating the unique keys, we then perform inner join on both dataframe to match the education index and poverty percentage of each city and of each year.

```python
def round_med(x):
    return np.round(np.median(x), 2)
edu_pov.groupby(by=["tahun_edu"]).agg(median_indeks=pd.NamedAgg("indeks_pendidikan", round_med),
                                      median_persentase=pd.NamedAgg("persentase_penduduk", round_med))
```

```python
diff_educ = pd.DataFrame()
diff_educ["nama_kabupaten_kota"] = edu_pov["nama_kabupaten_kota_edu"].unique()
diff_educ_idx = []
for city in diff_educ["nama_kabupaten_kota"]:
    locator = edu_pov["nama_kabupaten_kota_edu"]==city
    educ_city = edu_pov.loc[locator]
    educ_city = educ_city.sort_values("tahun_edu", ascending=True)
    initial = educ_city.iloc[0]["indeks_pendidikan"]
    final = educ_city.iloc[-1]["indeks_pendidikan"]
    diff_educ_idx.append((abs(final-initial)/initial)*100)
diff_educ["diff_educ_idx"] = diff_educ_idx
diff_educ = diff_educ.sort_values("diff_educ_idx", ascending=False)
color_map = [("green" if idx<5 else "red") for idx in range(diff_educ.shape[0])]

fig2, ax2 = plt.subplots(figsize=[8, 6])
_ = ax2.set_title("The Growth of Education Index Based On City or Regency\nin West Java, Indonesia")
_ = ax2.set_ylabel("Education Index")
_ = ax2.set_xlabel("City or Regency")
_ = ax2.bar(range(diff_educ.shape[0]), diff_educ["diff_educ_idx"], color=color_map)
for idx, y in enumerate(diff_educ["diff_educ_idx"]):
    _ = ax2.text(x=idx+0.025, y=y-0.2, s=str(round(y, 2))+"%", rotation="vertical", horizontalalignment="center",
                 verticalalignment="top", color="white")
_ = ax2.set_xticks(range(diff_educ.shape[0]), diff_educ["nama_kabupaten_kota"], rotation=90, fontsize=8)
_ = ax2.set_axisbelow(True)
_ = ax2.grid()
_ = fig2.tight_layout()
_ = fig2.savefig("./images/04_education_index_growth_jabar.jpg", dpi=600)
```

### Finding Correlation

```python
# removing outliers
[q1_educ, q3_educ] = edu_pov["indeks_pendidikan"].quantile(q=[0.25, 0.75]).values.tolist()
iqr_educ = q3_educ-q1_educ
upper_educ = q3_educ+(1.5*iqr_educ)
lower_educ = q1_educ-(1.5*iqr_educ)
query_educ = np.logical_and(edu_pov["indeks_pendidikan"]>=lower_educ, edu_pov["indeks_pendidikan"]<=upper_educ)

[q1_pov, q3_pov] = edu_pov["persentase_penduduk"].quantile(q=[0.25, 0.75]).values.tolist()
iqr_pov = q3_pov-q1_pov
upper_pov = q3_pov+(1.5*iqr_pov)
lower_pov = q1_pov-(1.5*iqr_pov)
query_pov = np.logical_and(edu_pov["persentase_penduduk"]>=lower_pov, edu_pov["persentase_penduduk"]<=upper_pov)

edu_pov_clean = edu_pov.loc[np.logical_and(query_educ, query_pov)].copy()

# calculating correlation
res_corrp = pearsonr(x=edu_pov_clean["indeks_pendidikan"], y=edu_pov_clean["persentase_penduduk"])
corrp = res_corrp.statistic
pvalue = res_corrp.pvalue
split_pvalue = str(pvalue).split("e")
print("Pearson's correlation: {}".format(round(corrp, 3)),
      "p-value: {}".format(round(float(split_pvalue[0]), 3)*(10**(int(split_pvalue[1])))),
      "p-value below 0.05: {}".format(pvalue<0.05),
      sep="\n")
```

```python
fig4, ax4 = plt.subplots(figsize=[8, 6])
_ = sns.scatterplot(data=edu_pov_clean, x="indeks_pendidikan", y="persentase_penduduk", hue="tahun_edu", ax=ax4, 
                    palette="viridis")
_ = ax4.set_title("Scatter Plot of Education Index vs. Poverty Percentage\nin West Java, Indonesia")
_ = ax4.set_ylabel("Poverty Percentage (%)")
_ = ax4.set_xlabel("Education Index")
_ = ax4.legend(title="Year")
_ = ax4.grid()
_ = ax4.set_axisbelow(True)
_ = fig4.tight_layout()
_ = fig4.savefig("./images/linear_rel_jabar.jpg", dpi=600)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
root_mse = lambda yt, yp: np.sqrt(mean_squared_error(y_pred=yp, y_true=yt))

x_val = edu_pov_clean["indeks_pendidikan"].values.reshape(-1, 1)
y_val = edu_pov_clean["persentase_penduduk"].values.reshape(-1, 1)

linreg = LinearRegression()
_ = linreg.fit(X=x_val, y=y_val)
y_pred = linreg.predict(X=x_val)
coef = linreg.coef_
intercept = linreg.intercept_
score = linreg.score(X=x_val, y=y_val)
print("MSE: {}".format(mean_squared_error(y_true=y_val, y_pred=y_pred)))
print("MAE: {}".format(mean_absolute_error(y_true=y_val, y_pred=y_pred)))
print("RMSE: {}".format(root_mse(y_val, y_pred)))
print("Coefficient: {}".format(round(coef.item(0), 3)),
      "Intercept: {}".format(round(intercept.item(0), 3)), "R-squared: {}".format(round(score, 3)),
      sep="\n")
```
